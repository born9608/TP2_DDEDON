```python
import pandas as pd
import numpy as np
# latlng_results (2)는 Google Map api를 이용하여 위경도 추출 이후 csv 파일 
data= pd.read_csv(r"C:\Users\rladn\Downloads\latlng_results (2).csv")
# "시군구-번지" 컬럼에서 "시군구"만 추출하여 새로운 컬럼 생성
# '시군구-번지' 컬럼 생성
data['시군구-번지'] = data['시군구']

# '시군구' 컬럼 수정
data['시군구'] = data['시군구'].str.split('-').str[0]

# 처음 5개 행 출력
data.head()
```

```python
# noise_ (2)는 진우님이 노이즈 반영해서 보간하신 데이터 가져온것
noise_data = pd.read_csv(r"C:\Users\rladn\Desktop\project_2_2\noise_ (2).csv")
noise_data ['전세가율'] = 100 * noise_data ['면적당보증금'] /noise_data ['면적당매매금']
noise_data.head()
```

```python
# Merging the two dataframes based on "시군구" column excluding '전세가율'
merged_data = pd.merge(data, noise_data[['시군구', '계약년월', '면적당보증금', '면적당매매금']], on='시군구', how='left')

# Displaying the merged data
merged_data.head()
```

```python
# Extract rows where '계약년월_y' has NaN values
nan_values = merged_data[merged_data['계약년월'].isna()]
# Remove rows where '계약년월_y' has NaN values
cleaned_data = merged_data.dropna(subset=['계약년월'])
cleaned_data['계약년월'] = cleaned_data['계약년월'].astype(int)
```

```python
cleaned_data= cleaned_data[['시군구','시군구-번지','계약년월','면적당보증금','면적당매매금','전세가율','lat','lng']]
# Renaming the columns as per the request
# cleaned_data = cleaned_data.rename(columns={
 #   '면적당보증금_y': '면적당보증금'
#})
cleaned_data.head()
```

```python
# Assigning a sequence number to rows with the same value in the '시군구-번지' column
cleaned_data['timestep'] = cleaned_data.groupby('시군구-번지').cumcount()

```

```python
# Reload the provided Excel file again for fresh data
cleaned_data.to_csv('cleaned_data.csv')
# 이 데이터로 forecasting_tool.py에서 추가 전처리 진행 
```
