```python
import pandas as pd



pd1 = pd.read_csv("시군구단위_noise_보간데이터_2_전세.csv")

pd2 = pd.read_csv("시군구단위_noise_보간데이터_2.csv")
```

```python
display(pd1)
display(pd2)
```

```python
pd1 = pd1[['시군구',	'계약년월',	'면적당보증금'	,'시']]
pd2 = pd2[['시군구',	'계약년월',	'면적당매매금'	,'시']]
```

```python
pd1
```

```python
merged_df = pd.merge(pd1, pd2, on=['시군구', '계약년월', '시'], suffixes=('면적당보증금', '면적당매매금'))
merged_df
```

```python
merged_df.info()
```

```python
merged_df.to_csv("시군구별매매전세_noise_보간.csv")
```

```python
merged_df['전세율'] =100*  merged_df['면적당보증금'] / merged_df['면적당매매금'] 
```

```python
def label_rent_rate(rate):
    if rate >= 90:
        return 1
    elif 70 <= rate < 90:
        return 2
    else:
        return 3
    
merged_df['레이블'] = merged_df['전세율'].apply(label_rent_rate)
merged_df
```
